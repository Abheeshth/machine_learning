```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
plt.style.use('seaborn')
```

```python
X,Y = make_regression(n_samples = 400,n_features = 1,n_informative = 1,noise = 1.0,random_state = 11)
Y = Y.reshape((-1,1))##this is beecause we try to make y as a matrix
X.shape ,Y.shape
```

```python
# normalize the data 
X = (X-X.mean())/X.std()
```

```python
#visualize
plt.figure()
plt.scatter(X,Y)
plt.title("normalize the data")
plt.show()
```

```python
ones  = np.ones((X.shape[0],1))
x = np.hstack((X,ones))
x.shape
```

```python
def pred(x,theta):
    return np.dot(x,theta)
```

```python
def getthetaclosedform(x,y):
    
    y = np.mat(y)   # here ae are changing y vector into matrix 
    firstpart = np.dot(x.T,x)
    secondpart = np.dot(x.T,y)
    
    theta = np.linalg.pinv(firstpart)*secondpart
    return theta
    
```

```python
theta = getthetaclosedform(x,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,pred(x,theta),color = 'red')
plt.title("normalize the data")
plt.show()
```
